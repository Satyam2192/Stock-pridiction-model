```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
df = pd.read_csv('../data/stock_price.csv', parse_dates=['日付け'])

# Basic statistics
print(df.describe())

# Plot stock price over time
plt.plot(df['日付け'], df['終値'])
plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
